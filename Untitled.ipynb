```python
import numpy as np
import pandas as pd
import os
import glob
```

```python
lego_data = {}

# use glob to get all the csv files 
# in the folder
path = os.getcwd()
csv_files = glob.glob(os.path.join(path, "data/*.csv"))
  
  
# loop over the list of csv files
for f in csv_files:
      
    # read the csv file
    df = pd.read_csv(f)
      
    # print the location and filename
    print('Location:', f)
    filename = f.split("\\")[-1]
    print('File Name:', filename)
    table_name = filename.split(".")[0]
    print(table_name)
    lego_data[table_name] = df
      
    # print the content
    print('Content:')
    display(df)
    print()
```

```python
lego_data['sets'].sort_values('num_parts', ascending=False)
```

```python
lego_data['inventories'].sort_values('set_num')
```

```python
count_sets = lego_data['sets'].groupby('theme_id').count().sort_values('set_num', ascending=False)
pd.merge(pd.DataFrame(count_sets), lego_data['themes'], left_on='theme_id', right_on='id')
```

```python
# Number of sets for each theme and year
grouped_sets = lego_data['sets'].groupby(['year', 'theme_id'])
pd.DataFrame(grouped_sets.size())
```

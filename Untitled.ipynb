```python
import numpy as np
import pandas as pd
import os
import glob
```

```python
lego_data = {}

# use glob to get all the csv files 
# in the folder
path = os.getcwd()
csv_files = glob.glob(os.path.join(path, "data/*.csv"))
  
  
# loop over the list of csv files
for f in csv_files:
      
    # read the csv file
    df = pd.read_csv(f)
      
    # print the location and filename
    print('Location:', f)
    filename = f.split("\\")[-1]
    print('File Name:', filename)
    table_name = filename.split(".")[0]
    print(table_name)
    lego_data[table_name] = df
      
    # print the content
    print('Content:')
    display(df)
    print()
```

```python
# Number of sets for ea year and theme
set_themes = pd.merge(
    lego_data['sets'][['set_num', 'theme_id', 'year']],
    lego_data['themes'][['id', 'name']],
    how='left',
    left_on='theme_id',
    right_on='id'
)

sets_year = pd.crosstab(set_themes['year'], set_themes['name'])
sets_year_cum = np.cumsum(sets_year)

sets_year.to_csv('data_new/sets_theme_year.csv')
sets_year_cum.to_csv('data_new/sets_theme_year_cum.csv')
```
